```python
# Загрузим все необходимые библиотеки
import os
import numpy as np
import pandas as pd

import plotly as py
import plotly.io as pio
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
#инициализация dataframe
FILE_ID = '1XV9IZ8cGArgCO9VkGId_IW2fewhg_tLN'  # ID таблицы
SHEET_NAME = 'Sheet1'  # имя листа (по умолчанию 'Sheet1')

url = f'https://drive.google.com/uc?export=download&id={FILE_ID}'

df = pd.read_csv(url)
print(df.shape)
df.head()
```

```python
df_genre = df[['Year','Genre','Global_Sales']].groupby(['Year', 'Genre'], as_index=False)['Global_Sales'].sum()
df_genre.shape
```

```python

df_genre = df_genre.dropna(subset=['Year'])  #удаляем возможные nan
df_genre['Year'] = df_genre['Year'].astype(int) #приводим год к int
print(df_genre.shape)
df_genre.head()
```

```python
fig = px.bar(
    df_genre,
    x='Year',
    y='Global_Sales',
    color='Genre',
    title='Динамика глобальных продаж по жанрам по годам',
    labels={'Global_Sales': 'Глобальные продажи', 'Year': 'Год'}
)

fig.update_layout(xaxis=dict(tickmode='linear', dtick=2))
fig.update_layout(barmode='group')
fig.show()
```

из график видно, что предпочтение по жанрам не практически не менялось со временем, практически во все времена в топе держатся Action, Misc и Sports, только в период начала зарождения игр лидерами были Puzzle, Platform и Shooter


```python
df_count_game = df[['Year','Name']].drop_duplicates()
df_count_game = df_count_game.dropna(subset=['Year'])  #удаляем возможные nan
df_count_game['Year'] = df_count_game['Year'].astype(int) #приводим год к int
print(df_count_game.shape)
df_count_game.head()
```

```python
df_count_game = df_count_game.groupby('Year', as_index=False)['Name'].count()
df_count_game.head()
```

```python
fig = px.bar(
    df_count_game,
    x='Year',
    y='Name',
    title='Динамика выпуска игр по годам',
    labels={'Name': 'Кол-во выпущенных игр', 'Year': 'Год'}
)

fig.update_layout(xaxis=dict(tickmode='linear', dtick=2))
fig.show()
```

```python
#Определи трех издателей, выпустивших наибольшее количество видеоигр.
#Изобрази количество выпущенных издателями видеоигр для каждой платформы на столбчатой диаграмме (можно использовать диаграмму с накоплением).

df_publisher = (
    df[['Publisher', 'Name']]
    .drop_duplicates()
    .groupby('Publisher', as_index=False)
    .agg(GameCount=('Name','count'))
    .sort_values('GameCount', ascending=False)
    .head(3)
)
df_publisher.head()

fig = px.bar(
    df_publisher,
    y='GameCount',
    x='Publisher',
    title='Лидеры рынка видео игр',
    labels={'GameCount': 'Кол-во выпущенных игр', 'Publisher': 'Производитель'}
)

fig.show()
```

```python
#Изобрази количество выпущенных издателями видеоигр для каждой платформы на столбчатой диаграмме (можно использовать диаграмму с накоплением).
df_publisher_platforms = (
    df[['Publisher', 'Platform', 'Name']]
    .drop_duplicates()
    .merge(df_publisher[['Publisher']], on='Publisher', how='inner')
    .groupby(['Publisher', 'Platform'], as_index=False)
    .agg(GameCount=('Name','count'))
)
df_publisher_platforms.head()
```

```python
fig = px.bar(
    df_publisher_platforms,
    y='GameCount',
    x='Publisher',
    color='Platform',
    title='Количество игр выпущенных производителем для каждой платформы',
    labels={'GameCount': 'Кол-во выпущенных игр', 'Publisher': 'Производитель'}
)

fig.show()
```
