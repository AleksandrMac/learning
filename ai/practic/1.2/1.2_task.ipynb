От тебя  требуется помимо создания самой модели получить ее точность на тестовой выборке. Выборки делить в соотношении 80% обучающая, 20% - тестовая.

Твоя задача с помощью Data Science предсказать заболевание паркинсона на ранней стадии, используя алгоритм машинного обучения **XGBoost** и библиотеку **sklearn** для нормализации признаков. Как это сделать? Тебя  придется самостоятельно изучить данный вопрос.

```python
# Загрузка данных по URL
url = "https://storage.yandexcloud.net/academy.ai/practica/parkinsons.data"

!wget -O parkinson.csv $url
```

```python
import pandas as pd

df = pd.read_csv("parkinson.csv")
```

```python
# Просмотр информация о данных
print(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

исходные данные в корректном состоянии, корректные числовые типы столбцов с отсутствующими null значениями

```python
# подготовка данных
# соберем все числовые значения в одном стоблце

# выбор числовых столбцов(целевых признаков)
X = df.select_dtypes(include=['float64', 'int64'])
Y = df['status']  # целевая переменная
```

```python
# нормализация данных
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Разделение данных на обучающую и тестовую выборки
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    X_scaled, Y, test_size=0.2, random_state=42
)
```

```python
print("Готово! Размеры:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
```

```python
from xgboost import XGBClassifier
bst = XGBClassifier(n_estimators=2, max_depth=2, learning_rate=1, objective='binary:logistic')
bst.fit(X_train, Y_train)
```

```python
preds = bst.predict(X_test)
print("Предсказания:", preds)

from sklearn.metrics import classification_report
print(classification_report(Y_test, preds))
```

По отчету классификации можно сказать, что модель показывает отличные результаты на тестовой выборке с точностью, полнотой и F1-мерой равными 1.00 для обоих классов (0 и 1). Это означает, что модель правильно классифицировала все примеры в тестовой выборке без ошибок.
